#### Getting numerical total value from the string value then use it to compute mean

```python
for i in range(0,1839):
    # check for nan value in sellerproductcount column
    if(df['sellerproductcount'][i]!= df['sellerproductcount'][i]):
        continue
    # split string to get the total results
    else:
        li = df['sellerproductcount'][i].split()
        if 'over' in li:
            li.remove('over')
        if '1-16' in li:
            li.remove('1-16')
        if 'of' in li:
            li.remove('of')
        df['sellerproductcount'][i] = li[0]
```

```python
print("number of NaN values for the column sellerproductcount :", df['sellerproductcount'].isnull().sum())
```

#### We can replace the missing values of the column 'sellerproductcount' with the mean of the column 'sellerproductcount'  using the method replace(). 

```python
# convert sellerproductcount column into numeric to calculate mean
df['sellerproductcount'] = pd.to_numeric(df['sellerproductcount'].str.replace(',', ''))
```

```python
mean=df['sellerproductcount'].mean()
mean1 = round(mean)
df['sellerproductcount'].replace(np.nan,mean1, inplace=True)
```

```python
df['sellerproductcount']
```

```python
print("number of NaN values for the column sellerproductcount :", df['sellerproductcount'].isnull().sum())
```

### Treating Nan values in the sellerratings column and parsing the ratings

#### Creating a new column 'Positive Ratings' from the seller ratings

```python
new = df['sellerratings'].str.split(" ", n = 1, expand = True)
```

```python
df['Positive Ratings'] = new[0]
for i in range(0,1839):
    #check for nan value
    if(new[0][i]!= new[0][i]):
        continue
    #split based on percentage to obtain only the rating
    pos = new[0][i].split('%')
    df['Positive Ratings'][i] = pos[0]
```

```python
for i in range(0,1839):
    #check for nan value
    if(new[1][i]!=new[1][i]):
        continue
    else:
        #split based on open brackets after which ratings is present
        li = new[1][i].split('(')
        num = li[1].split()
        df['sellerratings'][i] = num[0]
```

#### Replacing nan values for the columns sellerratings and Positive Ratings with the mean of their columns

```python
# convert sellerratings column into numeric to calculate mean
df['sellerratings'] = pd.to_numeric(df['sellerratings'].str.replace(',', ''))
```

```python
rate_mean=df['sellerratings'].mean()
rate_mean1 = round(rate_mean)
df['sellerratings'].replace(np.nan,rate_mean1, inplace=True)
```

```python
# convert Positive Ratings column into numeric to calculate mean
df['Positive Ratings'] = pd.to_numeric(df['Positive Ratings'].str.replace(',', ''))
```

```python
pos_rate_mean=df['Positive Ratings'].mean()
pos_rate_mean1 = round(pos_rate_mean)
df['Positive Ratings'].replace(np.nan,pos_rate_mean1, inplace=True)
```

```python
#display the data frame with the preprocessed data
df
```

#### Parsing the column sellerdetails to get the details about each seller

```python
# adding an email column
df['Email'] = 'NaN'
for i in range(0,1839):
    text = df['sellerdetails'][i]
    #check for nan value
    if(text!=text):
        continue
    emails = re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+", text)
    if(len(emails)!=0):
        df['Email'][i]=emails[0]
```

#### Excluding sellers from China

```python
#extracting country code
df['Country Code'] = 'NaN'
length = 1839
for i in range(0,length):
    address = df['businessaddress'][i]
    #check for nan
    if(address!=address):
        continue
    df['Country Code'][i] =  address[-2:]
```

### Taking into consideration Max % of negative seller ratings - last 12 months, last 90 and 30 days.
#### Weightage: 42% in last 12 months (as it has more correlation) + 33% in last 90 days + 25% in last 30 days.

```python
# creating a new column of Max % of negative seller ratings
df['Max % of negative seller ratings'] = round(0.42*df['Max % of negative seller ratings - last 12 months'] + 0.33*df['Max % of negative seller ratings - last 90 days'] + 0.25*df['Max % of negative seller ratings - last 30 days'])
```

## Proposed Formula:

### Best seller rate = [ (sellerproductcount / ( Countofsellerbrands + sellerratings ) * (PositiveRatings / 100) + Average( Hero Product 1 #ratings +  Hero Product 2 #ratings) ] / (Max % of negative seller ratings + 1 * 100)

### Creation of Average Hero Product Ratings

```python
df['Average Hero Product Ratings'] = round((df['Hero Product 1 #ratings']+ df['Hero Product 2 #ratings'])/2)
```

```python
df['Average Hero Product Ratings']
```

```python
df['Best Seller Rate'] = ( (df['sellerproductcount']/(df['sellerratings'] + df['Count of seller brands']))*(df['Positive Ratings']/100) + df['Average Hero Product Ratings'] ) / ((df['Max % of negative seller ratings'] + 1) *100)
```
